## Creating a function for implementing ridge regression and calculate the performance on validation dataset

```python
from sklearn.linear_model import Ridge
def ridge_regression(X, y, alpha,X_valid, Y_valid):
    #Fit the model
    print(alpha)
    ridgereg = Ridge(alpha=alpha,normalize=True)
    ridgereg.fit(X,y)
    y_pred = ridgereg.predict(X)
    
#     #Check if a plot is to be made for the entered alpha
#     if alpha in models_to_plot:
#         plt.subplot(models_to_plot[alpha])
#         plt.tight_layout()
#         plt.plot(data['x'],y_pred)
#         plt.plot(data['x'],data['y'],'.')
#         plt.title('Plot for alpha: %.3g'%alpha)
    
    #Finding RSS for Validation
    Y_valid_pred = ridgereg.predict(X_valid)
    rss_valid = sum((Y_valid_pred-Y_valid)**2)
    print('rss_valid')
    print(rss_valid)
        
    #Return the result in pre-defined format
    rss = sum((y_pred-y)**2)
    print(rss)
    ret = [rss]
    ret.extend([ridgereg.intercept_])
    ret.extend(ridgereg.coef_)
    return ret   
```
